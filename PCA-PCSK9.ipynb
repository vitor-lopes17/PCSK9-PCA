```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python

PCA_ChainA=pd.read_excel("pcsk9-gromacs-pca.xlsx")
```

```python
#Apo - CHAIN A


Apo_A=PCA_ChainA.loc[0:202]


plot = sns.jointplot(x="PC1", y="PC2", data=Apo_A, color = "silver", kind="kde")
plot.ax_marg_x.set_xlim(-1, 0.8)
plot.ax_marg_y.set_ylim(-0.6, 1)





fig1 = plt.gcf()
# plt.show()
# plt.draw()
fig1.savefig('Apo_A-PCA.pdf', dpi=600)
```
